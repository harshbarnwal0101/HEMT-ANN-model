```python
# 3. Log-transform the target to compress dynamic range
ml_df['Log_Current'] = np.log1p( ml_df['Drain_Current'] )
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import tensorflow as tf
from tensorflow.keras import layers, Sequential
```

```python
unique_drains = ml_df['Drain_Voltage'].unique()
train_drains = [6.0, 0.0, 1.0, 8.0, 9.0, 10.0,2.0,7.0]
test_drains = [3.0,4.0,5.0]

train_df = ml_df[ml_df['Drain_Voltage'].isin(train_drains)].reset_index(drop=True)
test_df  = ml_df[ml_df['Drain_Voltage'].isin(test_drains)].reset_index(drop=True)
```

```python
X_train = train_df[['Drain_Voltage', 'Gate_Voltage']].values
y_train = train_df['Log_Current'].values
X_test  = test_df[['Drain_Voltage', 'Gate_Voltage']].values
y_test_log = test_df['Log_Current'].values
y_test_actual = test_df['Drain_Current'].values
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
model = Sequential([
    layers.Dense(256, activation='relu', input_shape=(2,)),
    layers.Dense(512, activation='relu'),
    layers.Dense(256, activation='relu'),
    layers.Dense(1)
])
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# 8. Train model
history = model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    verbose=2
)
```

```python
# 10. Predict and evaluate
y_pred_log = model.predict(X_test).flatten()
y_pred_actual = np.expm1(y_pred_log)

# 10. Evaluate on actual current scale
mse = mean_squared_error(y_test_actual, y_pred_actual)
r2  = r2_score(y_test_actual, y_pred_actual)
mae = mean_absolute_error(y_test_actual, y_pred_actual)
print(f"Test MSE (A^2): {mse:.4e}, R²: {r2:.4f}, MAE (A): {mae:.4e}")

# Append the predicted current to the test_df
test_df['Predicted_Current'] = y_pred_actual
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.scatter(y_test_actual, y_pred_actual, alpha=0.6)
plt.plot([y_test_actual.min(), y_test_actual.max()], [y_test_actual.min(), y_test_actual.max()], 'r--')
plt.xlabel('Actual Drain Current (A)')
plt.ylabel('Predicted Drain Current (A)')
plt.title('Actual vs Predicted')
```

```python
residuals = y_test_actual - y_pred_actual
plt.subplot(1,2,2)
plt.scatter(y_pred_actual, residuals, alpha=0.6)
plt.axhline(0, linestyle='--')
plt.xlabel('Predicted (A)')
plt.ylabel('Residuals (A)')
plt.title('Residual Analysis')
plt.tight_layout()
plt.show()

# 12. Training & validation loss over epochs
plt.figure(figsize=(6,4))
plt.plot(history.history['loss'], label='Train Loss (log MSE)')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE on log-current)')
plt.title('Training History')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
for vg in sorted(test_df['Drain_Voltage'].unique()):  # Now using Drain_Voltage as the gate voltage
    mask = test_df['Drain_Voltage'] == vg
    Vg_vals = test_df.loc[mask, 'Gate_Voltage'].values  # Using Gate_Voltage as the Vds values
    Id_actual = test_df.loc[mask, 'Drain_Current'].values
    Id_pred = test_df.loc[mask, 'Predicted_Current'].values
    
    # Add origin point (0,0) - prepend to arrays
    Vg_vals = np.insert(Vg_vals, 0, 0.0)
    Id_actual = np.insert(Id_actual, 0, 0.0)
    Id_pred = np.insert(Id_pred, 0, 0.0)
    
    plt.figure(figsize=(6, 4))
    # Plot with swapped voltages, origin point included
    plt.plot(Vg_vals, Id_actual, 'o-', label='Actual')
    plt.plot(Vg_vals, Id_pred, 'x--', label='Predicted')
    plt.xlabel('Gate Voltage V$_G$ (V)')  # Label now represents Gate Voltage
    plt.ylabel('Drain Current I$_D$ (A)')
    plt.title(f'I$_D$ vs V$_{{G}}$ at V$_{{DS}}$ = {vg:.1f} V')  # Title reflects swapped voltages
    plt.legend()
    plt.grid(True, linestyle=':')
    plt.tight_layout()
    plt.show()
```
